### <mark style="background-color:teal; color:white"> Stacked Bar Chart 1 </mark>

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
data1 = [10, 20, 30, 40]
data2 = [5, 10, 5, 10]

fig, ax = plt.subplots()
ax.bar(x, data1)
ax.bar(x, data2)

fig.savefig('e16-1.png')
```

### <mark style="background-color:teal; color:white"> Stacked Bar Chart 2 </mark>

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
students = ['John', 'Doe', 'Mary', 'Bill']
x = [1,2,3,4]
math = [80, 90, 50, 80]
english = [60, 80, 40, 50]

ax.bar(students, math)
ax.bar(students, english, bottom=math)
fig.savefig('e16-2.png')
plt.show()
```

### <mark style="background-color:teal; color:white"> Grouped Bar Chart  </mark>

```python
import matplotlib.pyplot as plt
import numpy as np


label = ['Math', 'Physic', 'English', 'Algebra']
x = np.arange(len(label))
y1 = [100, 90, 100, 90]
y2 = [100, 100, 100, 90]
width = 0.35

fig, ax = plt.subplots()
rect1 = ax.bar(x-width/2, y1, width, label='John')
rect2 = ax.bar(x+width/2, y2, width, label='Doe')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
ax.set_xticks(x, label)
# plt.xticks(x, label)

ax.bar_label(rect1, padding=3)
ax.bar_label(rect2, padding=3)

fig.savefig('e16-3.png')
plt.show()
```
